### Naive Bayes

```python
# define X and y
X = bothdf[['age', 'rebounds', 'blocks', 'assists',
       'points','steals']]

y = bothdf['winner']
```

```python
# split X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
# train a logistic regression model on the training set
from sklearn.naive_bayes import GaussianNB

# instantiate model
nb = GaussianNB()

# fit model
nb.fit(X_train, y_train)
```

```python
# make class predictions for the testing set
#y_pred_class = nb.predict(X_test)
#y_pred_class

# 'age', 'rebounds', 'blocks', 'assists','points','steals'
new_data = np.array([[23,11.6,0.9,11.9,44.8,2.3]])


new_pred = nb.predict(new_data)
new_pred
```

```python
y_pred_class = nb.predict(X_test)
print(y_pred_class)
```

```python
# calculate accuracy
from sklearn import metrics
print(metrics.accuracy_score(y_test, y_pred_class))
```

```python
df2 = pd.concat([X_test, y_test], axis=1)
df2 = pd.concat([df2.reset_index(), pd.Series(y_pred_class)], axis=1)
df2.head(20)
```

```python
# IMPORTANT: first argument is true values, second argument is predicted values
# this produces a 2x2 numpy array (matrix)
print(metrics.confusion_matrix(y_test, y_pred_class))
```

```python
# save confusion matrix and slice into four pieces
confusion = metrics.confusion_matrix(y_test, y_pred_class)
print(confusion)
#[row, column]
TP = confusion[1, 1]
TN = confusion[0, 0]
FP = confusion[0, 1]
FN = confusion[1, 0]
```

```python
# Classification Accuracy: Overall, how often is the classifier correct?

# use float to perform true division, not integer division
print((TP + TN) / float(TP + TN + FP + FN))
print(metrics.accuracy_score(y_test, y_pred_class))
```

```python
# Classification Error: Overall, how often is the classifier incorrect?

# Also known as "Misclassification Rate"
classification_error = (FP + FN) / float(TP + TN + FP + FN)

print(classification_error)
print(1 - metrics.accuracy_score(y_test, y_pred_class))
```

```python
sensitivity = TP / float(FN + TP)

print(sensitivity)
print(metrics.recall_score(y_test, y_pred_class))
```

```python
specificity = TN / (TN + FP)

print(specificity)
```

```python
false_positive_rate = FP / float(TN + FP)

print(false_positive_rate)
print(1 - specificity)
```

```python
precision = TP / float(TP + FP)

print(precision)
print(metrics.precision_score(y_test, y_pred_class))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt



# IMPORTANT: first argument is true values, second argument is predicted probabilities

# we pass y_test and y_pred_prob

y_pred_prob = nb.predict_proba(X_test)[:, 1]

# we do not use y_pred_class, because it will give incorrect results without generating an error
# roc_curve returns 3 objects fpr, tpr, thresholds
# fpr: false positive rate
# tpr: true positive rate
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred_prob)

plt.plot(fpr, tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.rcParams['font.size'] = 12
plt.title('ROC curve for diabetes classifier')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.grid(True)
```

```python
# IMPORTANT: first argument is true values, second argument is predicted probabilities
print(metrics.roc_auc_score(y_test, y_pred_prob))
```

```python
from sklearn.neighbors import KNeighborsClassifier

# instantiate model
model = KNeighborsClassifier(n_neighbors=9)

# fit model
model.fit(X_train, y_train)
```

```python
# make class predictions for the testing set
y_pred_class = model.predict(X_test)
y_pred_class
```

```python
# calculate accuracy
from sklearn import metrics
print(metrics.accuracy_score(y_test, y_pred_class))
print(metrics.roc_auc_score(y_test, y_pred_class))
```

```python
k_values = [3, 5, 7, 9]
for k in k_values:
    # instantiate model
    model = KNeighborsClassifier(n_neighbors=k)

    # fit model
    model.fit(X_train, y_train)
    print("Neighbors: ", k)
    print("Accuracy: ", metrics.accuracy_score(y_test, y_pred_class))
```
